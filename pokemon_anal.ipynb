Battle 데이터에서 포켓몬 이름과 스탯을 딕셔너리화

```python
def change_data(data):
    data['First_pokemon_stats'] = data.First_pokemon.map(dictionary_stats)
    data['Second_pokemon_stats'] = data.Second_pokemon.map(dictionary_stats)
    data['First_pokemon'] = data.First_pokemon.map(dictionary_type)
    data['Second_pokemon'] = data.Second_pokemon.map(dictionary_type)
    return data
```

```python
df_train = change_data(df_battle)
df_train
```

```python
def calc_stats(data):
    stats = ['HP_d', 'Attack_d', 'Defense_d', 'Speed_d', 'Sp. Atk_d', 'Sp. Def_d', 'Legendary_d']
    d_list = []
    
    for i in data.itertuples():
        d_list.append(np.array(i.First_pokemon_stats)-np.array(i.Second_pokemon_stats))
        
    df_new_stats = pd.DataFrame(d_list, columns=stats)
    data = pd.concat([data, df_new_stats], axis=1)
    data.drop(['First_pokemon_stats', 'Second_pokemon_stats'], axis=1, inplace=True)
    return data
```

```python
df_train = calc_stats(df_train)
df_train
```

```python
def calc_effective(data):

    
    dictionary_effective = {'Normal': [],
                           'Fighting': ['Normal', 'Rock', 'Steel', 'Ice', 'Dark'],
                           'Flying': ['Fighting', 'Bug', 'Grass'],
                           'Poison': ['Grass', 'Fairy'],
                           'Ground': ['Poison', 'Rock', 'Steel', 'Fire', 'Electric'],
                           'Rock': ['Flying', 'Bug', 'Fire', 'Ice'],
                           'Bug': ['Grass', 'Psychic', 'Dark'],
                           'Ghost': ['Ghost', 'Psychic'],
                           'Steel': ['Rock', 'Ice', 'Fairy'],
                           'Fire': ['Bug', 'Steel', 'Grass', 'Ice'],
                           'Water': ['Ground', 'Rock', 'Fire'],
                           'Grass': ['Ground', 'Rock', 'Water'],
                           'Electric': ['Flying', 'Water'],
                           'Psychic': ['Fighting', 'Poison'],
                           'Ice': ['Flying', 'Ground', 'Grass', 'Dragon'],
                           'Dragon': ['Dragon'],
                           'Dark': ['Ghost', 'Psychic'],
                           'Fairy': ['Fighting', 'Dragon', 'Dark'],
                           'Not': []}
    dictionary_not_effective = {'Normal': ['Rock', 'Steel'],
                               'Fighting': ['Flying', 'Poison', 'Bug', 'Psychic', 'Fairy'],
                               'Flying': ['Rock', 'Steel', 'Electric'],
                               'Poison': ['Poison', 'Rock', 'Ground', 'Ghost'],
                               'Ground': ['Bug', 'Grass'],
                               'Rock': ['Fighting', 'Ground', 'Steel'],
                               'Bug': ['Fighting', 'Flying', 'Poison', 'Ghost', 'Steel', 'Fire', 'Fairy'],
                               'Ghost': ['Dark'],
                               'Steel': ['Steel', 'Fire', 'Water', 'Electric'],
                               'Fire': ['Rock', 'Fire', 'Water', 'Dragon'],
                               'Water': ['Water', 'Grass', 'Dragon'],
                               'Grass': ['Flying', 'Poison', 'Bug', 'Steel', 'Fire', 'Grass', 'Dragon'],
                               'Electric': ['Grass', 'Electric', 'Dragon'],
                               'Psychic': ['Steel', 'Psychic'],
                               'Ice': ['Steel', 'Fire', 'Water', 'Psychic'],
                               'Dragon': ['Steel'],
                               'Dark': ['Fighting', 'Dark', 'Fairy'],
                               'Fairy': ['Posion', 'Steel', 'Fire'],
                               'Not': []}
    dictionay_never_effective = {'Normal': ['Ghost'],
                          'Fighting': ['Ghost'],
                          'Flying': [],
                          'Poison': ['Steel'],
                          'Ground': ['Flying'],
                          'Rock': [],
                          'Bug': [],
                          'Ghost': ['Normal'],
                          'Steel': [],
                          'Fire': [],
                          'Water': [],
                          'Grass': [],
                          'Electric': ['Ground'],
                          'Psychic': ['Dark'],
                          'Ice': [],
                          'Dragon': ['Fairy'],
                          'Dark': [],
                          'Fairy': [],
                          'Not': []}
    player1_type1 = []
    player1_type2 = []
    player2_type1 = []
    player2_type2 = []
    for line in data.itertuples():
        element_type = [[1, 1], [1, 1]]
        for i in range(0, 2):
            for j in range(0, 2):
                if line.Second_pokemon[j] in dictionary_effective.get(line.First_pokemon[i]):
                    element_type[0][i] *= 2
                if line.Second_pokemon[j] in dictionary_not_effective.get(line.First_pokemon[i]):
                    element_type[0][i] /= 2
                if line.Second_pokemon[j] in dictionay_never_effective.get(line.First_pokemon[i]):
                    element_type[0][i] *= 0
                if line.First_pokemon[j] in dictionary_effective.get(line.Second_pokemon[i]):
                    element_type[1][i] *= 2
                if line.First_pokemon[j] in dictionary_not_effective.get(line.Second_pokemon[i]):
                    element_type[1][i] /= 2
                if line.First_pokemon[j] in dictionay_never_effective.get(line.Second_pokemon[i]):
                    element_type[1][i] *= 0
        
        player1_type1.append(element_type[0][0])
        player1_type2.append(element_type[0][1])
        player2_type1.append(element_type[1][0])
        player2_type2.append(element_type[1][1])
        
    data = data.assign(player1_type1_r=player1_type1, player1_type2_r = player1_type2, player2_type1_r=player2_type1, \
                      player2_type2_r=player2_type2)
    data = data.drop(['First_pokemon', 'Second_pokemon'], axis=1)
    
    return data
```

```python
df_train = calc_effective(df_train)
df_train
```

```python
y_all_for_train = df_train['Winner']
x_all_for_train = df_train.drop('Winner', axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_all_for_train, y_all_for_train, random_state=123)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
```

```python
def predict(model):
    prediction = model.fit(x_train, y_train).predict(x_test)
    print("Accuracy : ", accuracy_score(prediction, y_test))
    return prediction
```

```python
log_pred = predict(LogisticRegression()) # Logistic Regression
svm_pred = predict(LinearSVC()) # Linear SVM
knn_pred_4 = predict(KNeighborsClassifier(n_neighbors = 4)) # kNN
rf_pred = predict(RandomForestClassifier(n_estimators=100)) # Random Forest
nb_pred = predict(GaussianNB()) # Navie Bayes
dt_pred = predict(DecisionTreeClassifier())
```
